```python
import matplotlib.pyplot as plt

# データ
x = [0, 0.00043962, 0.0008772, 0.0013086, 0.00175848, 0.0021981, 0.00263742, 0.00307572]
y = [0, 0.03, 0.06, 0.09, 0.12, 0.15, 0.18, 0.21]

# プロット
plt.figure(figsize=(8, 6))  # グラフのサイズを設定
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Square lattice')  # プロットの設定
plt.title('Square lattice Bending')  # グラフのタイトル
plt.xlabel('Strain : $\epsilon$[-]')  # X軸のラベル
plt.ylabel('Stress : $\sigma$[MPa]')  # Y軸のラベル
plt.grid(True)  # グリッドの表示
plt.legend()  # 凡例の表示
plt.tight_layout()  # レイアウトの調整（必要に応じて）

# グラフの表示
plt.show()
```

```python
import plotly.graph_objects as go

# データ
x = [0, 0.00043962, 0.0008772, 0.0013086, 0.00175848, 0.0021981, 0.00263742, 0.00307572]
y = [0, 0.03, 0.06, 0.09, 0.12, 0.15, 0.18, 0.21]

# Figureオブジェクトを作成
fig = go.Figure()

# 折れ線グラフを追加
fig.add_trace(go.Scatter(x=x, y=y, mode='lines+markers', name='Square lattice Bending'))

# レイアウト設定
fig.update_layout(
    title='Square lattice Bending',  # グラフのタイトル
    xaxis_title='Strain : $\epsilon$[-]',  # X軸のラベル
    yaxis_title='Stress : $\sigma$[MPa]',  # Y軸のラベル
    showlegend=True,  # 凡例を表示
    xaxis=dict(showgrid=True),  # X軸のグリッドを表示
    yaxis=dict(showgrid=True),  # Y軸のグリッドを表示
)

# グラフの表示
fig.show()
```
